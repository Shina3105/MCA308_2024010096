**Data Visualization**

```python
sns.histplot(data['price'],bins=50,kde=True)
plt.title("Distribution of Price")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='neighbourhood_group', data=data, palette='pastel')
plt.title('Number of Listings by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='room_type', data=data, palette='Set2')
plt.title('Distribution of Room Types')
plt.show()
```

```python
sns.boxplot(x='room_type', y='price', data=data)
plt.title('Price Distribution by Room Type')
plt.show()
```

```python
plt.figure(figsize=(13,8))
sns.scatterplot(x='longitude', y='latitude', hue='price', data=data, palette='coolwarm')
plt.title('Price Distribution Across NYC')
plt.show()
```

```python
sns.scatterplot(x='number_of_reviews', y='price', data=data, alpha=0.6)
plt.title('Price vs. Number of Reviews')
plt.show()
```

```python
sns.pairplot(data[['price', 'minimum_nights', 'number_of_reviews', 'availability_365']])
plt.show()
```

```python
cols = [
    'latitude', 'longitude', 'price', 'minimum_nights',
    'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365'
]

sns.heatmap(data[cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Selected Features)')
plt.show()
```
