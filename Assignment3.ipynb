```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("AWSales.csv")
group_means = df.groupby("BikeBuyer")["AvgMonthSpend"].mean()
plt.bar(group_means.index, group_means.values, color=['pink','green'])
plt.xticks([0,1], ["Non-Buyers", "Buyers"])
plt.ylabel("Average Monthly Spend")
plt.show()
```
