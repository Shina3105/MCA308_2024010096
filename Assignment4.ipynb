```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(45)
data=np.random.randint(1,200,size=(100,30))
```

```python
data
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df = df.map(lambda x: np.nan if 10 <= x <= 60 else x)
```

```python
df
```

```python
df.isna().sum()
```

```python
df = df.apply(lambda col: col.fillna(col.mean()), axis=0)
```

```python
df
```

```python
plt.figure(figsize=(35, 50))
sns.heatmap(df, cmap="magma",annot=True)
plt.title("Heatmap of Dataset")
plt.show()
```

```python
corr_matrix = df.corr()
low_corr_cols = (corr_matrix <= 0.7).all(axis=0).sum()
low_corr_cols
```

```python
normalized_df = df.apply(lambda col: 10 * (col - col.min()) / (col.max() - col.min()))
```

```python
binary_df = normalized_df.map(lambda x: 0 if x <= 5 else 1)
```

```python
normalized_df.hist(bins=20, figsize=(15, 10))
plt.suptitle("Normalized Dataset", fontsize=14)
plt.show()
```

```python
for col in df.columns:
    plt.figure(figsize=(4,2))
    plt.bar(col , df[:-1])
    plt.show()
```
